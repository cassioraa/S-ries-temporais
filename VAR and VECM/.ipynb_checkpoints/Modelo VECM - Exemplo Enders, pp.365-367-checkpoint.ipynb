# ESTIMANDO MODELO VECM - ABORDAGEM ENGLE-GRANGER

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Carregando dados
dados = pd.read_csv('~/Documents/Estágio Docência/Aula VAR/Python/Coint6.38140536.csv')
dados.head()
```

```python
# Figura 6.2, p. 365
plt.plot(dados["y"].values, label="y")
plt.plot(dados["z"].values, label="z")
plt.plot(dados["w"].values, label="w")
plt.axhline(0, color="black")
plt.legend()
plt.show()
```

```python
#==================================
# Verificando se há cointegração
#==================================

# Regressão y_t = b0 + b1 z_t + b2 w_t
y = dados["y"]
X = dados[["z","w"]]
X = sm.add_constant(X) # Adiciona coluna de 1's para estimar b0

modelo = sm.OLS(y,X).fit()
e_y = modelo.resid
modelo.summary()
```

```python
# Regressão z_t = b0 + b1 y_t + b2 w_t
z = dados["z"]
X = dados[["y","w"]]
X = sm.add_constant(X) # Adiciona coluna de 1's para estimar b0

modelo = sm.OLS(z,X).fit()
e_z = modelo.resid
modelo.summary()
```

```python
# Regressão w_t = b0 + b1 y_t + b2 z_t
w = dados["w"]
X = dados[["y","z"]]
X = sm.add_constant(X) # Adiciona coluna de 1's para estimar b0

modelo = sm.OLS(w,X).fit()
e_w = modelo.resid
modelo.summary()
```

```python
#==================================
# Análise dos resíduos das 3 eqs.
#==================================

# Esses resíduos são estacionários? Aparente pela inspeção visual sim. Fazer um teste ADF
plt.figure(figsize=(10,3))
plt.plot(e_y, color="black")
plt.plot(e_z, color="blue")
plt.plot(e_w, color="red")
plt.show()
```

```python
# Montando a variável dependente

Dy = dados['y'].diff().dropna()
Dz = dados['z'].diff().dropna()
Dw = dados['w'].diff().dropna()
# Montando os regressores da primeira equação do VECM


e_w1 = e_w[1:].shift().dropna().values
Dy1 = Dy.shift().dropna().values
Dz1 = Dz.shift().dropna().values
Dw1 = Dw.shift().dropna().values



X = pd.DataFrame(
    np.vstack((e_w1, Dy1, Dz1, Dw1)).T
)

X = sm.add_constant(X)
Dy = Dy[1:]
Dy.index = X.index
```

```python
# == Estima a equação 1 usando a matriz de regressores == #

# fit(cov_type='HAC',cov_kwds={'maxlags':1})
# para estimar erro-padrão consistente consistente com Autocorrelação e heterocedasticidade

eq1 = sm.OLS(Dy,X).fit()
eq1.summary()
```

```python
# == Estima a equação 2 usando a matriz de regressores == #

# fit(cov_type='HAC',cov_kwds={'maxlags':1})
# para estimar erro-padrão consistente consistente com Autocorrelação e heterocedasticidade
Dz = Dz[1:]
Dz.index = X.index
eq2 = sm.OLS(Dz,X).fit()
eq2.summary()
```

```python
# == Estima a equação 3 usando a matriz de regressores == #

# fit(cov_type='HAC',cov_kwds={'maxlags':1})
# para estimar erro-padrão consistente consistente com Autocorrelação e heterocedasticidade
Dw = Dw[1:]
Dw.index = X.index
eq3 = sm.OLS(Dw,X).fit()
eq3.summary()
```
